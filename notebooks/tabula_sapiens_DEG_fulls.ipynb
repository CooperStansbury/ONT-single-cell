```python
import pandas as pd
import numpy as np
import anndata as an
import os
import sys
import scipy
import gget

import scanpy as sc
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fpath = "/nfs/turbo/umms-indikar/shared/projects/DGC/data/tabula_sapiens/TabulaSapiens.h5ad"

adata = sc.read_h5ad(fpath)
adata
```

```python
sc.tl.rank_genes_groups(adata, 
                        groupby="free_annotation",
                        method='wilcoxon',
                        corr_method='benjamini-hochberg',
                        pts=True,
                        key_added='cell_ontology_class',
                       )

deg = sc.get.rank_genes_groups_df(adata, 
                                  group=None,
                                  key='cell_ontology_class')

outpath = "/nfs/turbo/umms-indikar/shared/projects/HSC/data/tabula_sapiens/DEG/cell_ontology_class.csv"
deg.to_csv(outpath, index=False)
deg.head()
```

```python
sc.tl.rank_genes_groups(adata, 
                        groupby="free_annotation",
                        method='wilcoxon',
                        corr_method='benjamini-hochberg',
                        pts=True,
                        key_added='free_annotation_deg',
                       )

deg = sc.get.rank_genes_groups_df(adata, 
                                  group=None,
                                  key='free_annotation_deg')

outpath = "/nfs/turbo/umms-indikar/shared/projects/HSC/data/tabula_sapiens/DEG/free_annotation_deg.csv"
deg.to_csv(outpath, index=False)
deg.head()
```
