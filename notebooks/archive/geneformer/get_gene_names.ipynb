```python
import pickle
import pyranges as pr
import pandas as pd
```

```python
fpath = "/nfs/turbo/umms-indikar/shared/projects/geneformer/token_dictionary.pkl"

def open_pickle(file_path):
    with open(file_path, 'rb') as file:
        loaded_object = pickle.load(file)
    return loaded_object

token_dict = open_pickle(fpath)
df = pd.DataFrame.from_dict(token_dict, orient='index')
df = df.reset_index()
df.columns = ['gene_id', 'token_id']
print(f"{df.shape=}")
df.head()
```

```python
gtf_path = "/scratch/indikar_root/indikar1/cstansbu/HSC/references/geneTable.csv"
gtf = pd.read_csv(gtf_path)
gtf = gtf[gtf['Feature'] == 'gene']
print(f"{gtf.shape=}")
gtf.head()
```

```python
# merge it
df = pd.merge(df, gtf, 
              how='left',
              left_on='gene_id',
              right_on='gene_id',
             )

print(f"{df.shape=}")
outpath = "/nfs/turbo/umms-indikar/shared/projects/geneformer/token_mappings.csv"
df.to_csv(outpath, index=False)
df.head()
```
