```python
import velocyto as vcy
import matplotlib.pyplot as plt
import numpy as np
```

```python
fpath = "/scratch/indikar_root/indikar1/cstansbu/velo/output/test_LZSZM.loom"

vlm = vcy.VelocytoLoom(fpath)

vlm.normalize("S", size=True, log=True)

vlm.plot_fractions()
```

```python
plt.scatter(vlm.initial_cell_size, vlm.initial_Ucell_size, alpha=0.5, s=5)
plt.axvline(250, c="r", lw=1)
plt.axvline(np.percentile(vlm.initial_cell_size, 8), c="k", lw=1)
plt.axhline(150, c="r", lw=1)
plt.axhline(np.percentile(vlm.initial_Ucell_size, 8), c="k", lw=1)
plt.xlabel("spliced"); plt.ylabel("unspliced")

# plt.yscale('log')
# plt.xscale('log')
```
