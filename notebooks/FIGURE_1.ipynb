```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fpath = "../resources/hsc_flow_CD34.csv"
df = pd.read_csv(fpath)

df['label'] = df['comment'].str.lower()
rep_label = "reprogram"
df['label'] = df['label'].replace("low", rep_label)
df['label'] = df['label'].replace("high", rep_label)

df['label'] = df['label'].replace("no lenti no dox", "lenti-/dox-")
df['label'] = df['label'].replace("no dox", "dox-")

df['condition'] = df['condition'].replace("control", 'Control')
df
```

```python
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.figsize'] = 2, 2.25

sns.barplot(
    data=df,
    x='label',
    y='percent',
    ec='k',  # Edge color black
    err_kws={'color': 'black','linewidth': 1.5},
    capsize=0.1  # Smaller error bar caps
)

plt.ylabel(r"% CD34$^+$")
plt.xlabel("")

plt.gca().tick_params(axis='x', rotation=90)


sns.despine()
```
