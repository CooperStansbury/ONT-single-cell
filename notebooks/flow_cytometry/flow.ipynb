```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.axes_grid1.axes_divider import make_axes_locatable

import flowkit as fk


"""WARNING: disabling future warnings"""
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

# FH6

```python
dpath = "/nfs/turbo/umms-indikar/shared/projects/HSC/data/FLOW/FH6/"

df_list = []

for f in os.listdir(dpath):
    label = " ".join(f.replace(".fcs", "").split(" ")[1:])
    print(label)
    slide = f.replace(".fcs", "").split(" ")[0]
    fpath = f"{dpath}{f}"
    sample = fk.Sample(fpath,ignore_offset_error=True)
    df = sample.as_dataframe(source='raw')
    df['sample'] = label
    df['slide'] = slide
    df_list.append(df)
    # break

df = pd.concat(df_list)
print(f"{df.shape=}")
df.head()
```

```python
sample_list = ['Delay 12 Sep G,R,B', 'Delay 12 Sep R,G,B']
x = ('FL03-A', 'EGFP-A')
y = ('FL15-A', 'mCherry-A')

def plot_flow_scatter(df, x, y, ax=None, 
                      scatter_kws=False,
                      hist_kws=False,
                      kde_kws=False):
    """A function to plot flow results as a scatterplot """
    if ax is None:
        _, ax = plt.subplots()

    if not scatter_kws is None:
        sns.scatterplot(data=df, 
                        x=x,
                        y=y,
                        ax=ax,
                        **scatter_kws,
               )
    if not hist_kws is None:
        sns.histplot(data=pdf, 
                     x=x,
                     y=y,
                     ax=ax,
                     **hist_kws,
                )
    if not kde_kws is None:
        sns.kdeplot(data=pdf, 
                    x=x,
                    y=y, 
                    ax=ax,
                    **kde_kws,
           )

    return ax

            
    
    

for i, sample in enumerate(sample_list):

    pdf = df[df['sample'] == sample]

    plt.rcParams['figure.dpi'] = 300
    plt.rcParams['figure.figsize'] = 4, 4
    
    ax = plot_flow_scatter(pdf, x, y, 
                      scatter_kws={'s' : 0.5, 
                                   'color' : 'k', 
                                   'alpha' : 0.75, 
                                   'ec':'k', 
                                   'zorder' : 3},
                      hist_kws={'bins' : 50, 
                                'log_scale' : True,
                                'alpha' : 0.75, 
                                'cmap' : 'hot_r',
                                'zorder' : 2},
                      kde_kws={'levels' : 5,
                               'log_scale' : True,
                               'color' : 'k',
                               'linewidths' : 0.5,
                               'zorder' : 4,
                              })

    ax.set_title(sample)
    ax.set_xlabel(x[1])
    ax.set_ylabel(y[1])
    
    ax.grid(True, 
            which='major',
            zorder=0,
            c='k', 
            alpha=0.5,
            lw=0.1,
    )
    sns.despine()
    plt.show()
```
